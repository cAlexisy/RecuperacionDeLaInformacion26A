- Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

### Paso 1: Calcular la matriz TF-IDF para el corpus `data/01_corpus_turismo_500.txt`

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Paso 1: Guardar la ruta del archivo en una variable
# Se usa la ruta relativa que apunta a la carpeta data
ruta = '../data/01_corpus_turismo_500.txt'

# Paso 2: Abrir el archivo de texto
# Se usa 'r' para leer y utf-8 para que reconozca las tildes
archivo = open(ruta, 'r', encoding='utf-8')

# Paso 3: Leer todas las lineas y guardarlas en una lista
read = archivo.readlines()

# Paso 4: Cerrar el archivo despues de leerlo para liberar memoria
archivo.close()

# Paso 5: Limpiar las lineas eliminando espacios en blanco extra
corpus = []
for l in read:
    texto_limpio = l.strip()
    if texto_limpio != "":
        corpus.append(texto_limpio)

# Paso 6: Crear el objeto vectorizador para TF-IDF
# Este objeto hara todos los calculos matematicos por nosotros
vectorizador = TfidfVectorizer()

# Paso 7: Aplicar el calculo a nuestro corpus de texto
# El resultado es una matriz con los pesos de cada palabra
matriz_tfidf = vectorizador.fit_transform(corpus)

# Paso 8: Obtener la lista de todas las palabras que encontro el programa
nombres_columnas = vectorizador.get_feature_names_out()

# Paso 9: Convertir los resultados a una tabla de Pandas para que sea legible
df_resultado = pd.DataFrame(matriz_tfidf.toarray(), columns=nombres_columnas)

# Paso 10: Imprimir las primeras filas para verificar que todo este bien
print(df_resultado.head())
```

### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
import os

# Ruta donde estan los 1000 archivos
ruta = '../data/gutenberg/data'

corpus_1000 = []
nombres_1000 = []

# Obtenemos la lista de los 1000 archivos
archivos_listos = [f for f in os.listdir(ruta) if f.endswith('.txt')]

# Tomamos solo los primeros 1000 por si acaso hay mas
archivos_listos = archivos_listos[:1000]

for nombre in archivos_listos:
    ruta_archivo = os.path.join(ruta, nombre)
    
    # Leemos cada libro
    # 'errors=ignore' es vital porque algunos libros tienen simbolos que Python no entiende
    with open(ruta_archivo, 'r', encoding='utf-8', errors='ignore') as f:
        texto = f.read()
        corpus_1000.append(texto)
        nombres_1000.append(nombre)

print("Corpus con " + str(len(corpus_1000)) + " documentos.")
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Paso 1: Configurar el vectorizador
# max_features=1500 solo use las 1500 palabras mas importantes
vectorizador_1000 = TfidfVectorizer(max_features=1500)

# Paso 2: Calcular la matriz TF-IDF usando el corpus de los 1000 libros
# fit_transform hace todo el trabajo matematico de contar y ponderar las palabras
matriz_1000 = vectorizador_1000.fit_transform(corpus_1000)

# Paso 3: Obtener los nombres de las palabras (columnas)
vocabulario_1000 = vectorizador_1000.get_feature_names_out()

# Paso 4: Crear la tabla (DataFrame) para ver los resultados
# Usamos los nombres de los libros (nombres_1000) para identificar las filas
df_gutenberg_1000 = pd.DataFrame(
    matriz_1000.toarray(), 
    columns=vocabulario_1000, 
    index=nombres_1000
)

# Paso 5: Mostrar cuantas filas y columnas tiene la matriz final
print("Total de libros (filas): " + str(df_gutenberg_1000.shape[0]))
print("Total de palabras (columnas): " + str(df_gutenberg_1000.shape[1]))

# Paso 6: Ver las primeras 5 filas de la tabla
print(df_gutenberg_1000.head())
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
from sklearn.metrics.pairwise import linear_kernel

def buscar(consulta, n=5,):
    """
    Busca los libros más relevantes en el corpus Gutenberg 1000.
    """
    # 1. Transformar la frase de búsqueda al mismo vocabulario de los 1000 libros
    query_tfidf = vectorizador_1000.transform([consulta])
    # 2. Calcula la similitud entre la consulta y TODOS los libros a la vez
    similitudes = linear_kernel(query_tfidf, matriz_1000).flatten()
    # 3. Crear una serie de Pandas para organizar los resultados
    ranking = pd.Series(similitudes, index=nombres_1000).sort_values(ascending=False)
    # 4. Retornar los top 'n' resultados
    print(f"Top {n} libros más relevantes para: '{consulta}'")
    return ranking.head(n)

# --- PRUEBA DEL BUSCADOR ---
buscar("astronomy and stars", n=10)
```
